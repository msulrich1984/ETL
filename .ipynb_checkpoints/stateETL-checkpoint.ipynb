```python
import pandas as pd
from sqlalchemy import create_engine  #Store CSV into DataFrame
import json
from Resources import configmu
import pprint
import pymysql
from pandas.io import sql
```

```python
csv_file = "population.csv"
pop_data_df = pd.read_csv(csv_file)
pop_data_df.columns
```

```python
state_json = json.load(open('state_sociological.json'))
```

```python
state_json
```

```python
state_df = pd.DataFrame(state_json["data"])
state_df
```

```python
renamed_state_df = pop_data_df.rename(columns={"NAME":"state", "STATE":"index"})
renamed_state_df
```

```python
pd.merge(state_df, renamed_state_df,)
merge_df = pd.merge(state_df,renamed_state_df[['state','POPESTIMATE2010','POPESTIMATE2016','NRANK_POPEST2016','NRANK_NPCHG2010']], how='left')
merge_df
```

```python
def popChange (row):
    num1 = row['POPESTIMATE2016']
    num2 = row['POPESTIMATE2010']
    change = num1 - num2
    
    return change
```

```python
merge_df['TOTALPOPCHG'] = merge_df.apply (lambda row: popChange (row),axis=1)

merge_df.head()
```

```python
def rankChange (row):
    num1 = int(row['NRANK_POPEST2016'])
    num2 = int(row['NRANK_NPCHG2010'])
    change = num1 - num2
    
    return change
```

```python
merge_df['TOTALRNKCHANGE'] = merge_df.apply (lambda row: rankChange (row),axis=1)

merge_df.head()
```

```python
renamed_merge = merge_df.rename(columns={"percent_above_poverty_rate":"Percent above Poverty Rate", "percent_educational_attainment":"Percent Bachelor's Degree", "percent_non_religious":"Percent Non-Religious", "percent_peace_index":"Percent Peace Index","state":"State", "POPESTIMATE2010":"Population Estimate 2010", "POPESTIMATE2016":"Population Estimate 2016", "NRANK_POPEST2016":"2016 Rank","NRANK_NPCHG2010":"2010 Rank","TOTALPOPCHG":"Total Population Change","TOTALRNKCHANGE":"Total Rank Change"})
renamed_merge.head() 
```

```python
state_merge=renamed_merge.set_index('State')
state_merge.head()
```
